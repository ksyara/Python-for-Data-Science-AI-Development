```python
%%sql 
select count(*) as number_of_crimes, COMMUNITY_AREA_NUMBER
from chicago_crime_data
where COMMUNITY_AREA_NUMBER is not null
group by COMMUNITY_AREA_NUMBER
order by number_of_crimes desc limit 1
```

```python
%%sql 
select hardship_index, community_area_name
from chicago_census_data
where hardship_index = (select max(hardship_index) as max_hardship_index from chicago_census_data)
```

```python
%%sql 
select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME from chicago_census_data 
where COMMUNITY_AREA_NUMBER in (select community_area_number 
                                from chicago_crime_data 
                                order by community_area_number desc nulls last limit 1)
```

```python
%%sql 
select community_area_number, count(id) as crime_count from chicago_crime_data
group by community_area_number
order by crime_count desc limit 1


```

```python
%%sql 
select ce.community_area_number, count(case_number) as crime_count, community_area_name
from chicago_crime_data cr, chicago_census_data ce
where ce.community_area_number = cr.community_area_number
group by ce.community_area_number, community_area_name
order by crime_count desc limit 1

```
